# 3. Introduction


Thermochemistry, the primary focus of this experiment, is defined as the branch of thermodynamics concerned with the heat changes that occur during chemical reactions (Atkins, De Paula, & Keeler, 2025). Chemical reactions are categorized based on their physical and chemical transformations, but at its core, thermochemistry aims to quantitatively analyze these reactions by measuring key variables such as energy changes, the intrinsic properties of reactants and products, and their environmental conditions (Ruscic & Bross, 2019). Calorimetry serves as a fundamental technique in this analysis, providing sensitive and precise measurements of heat changes, which directly relate to bond energies and reaction mechanisms (The relevance of calorimetry to physical organic chemistry).

To effectively use calorimetry, it is essential to understand the First Law of Thermodynamics—mathematically expressed as the conservation of internal energy within an isolated system—and its simplifications under constant pressure and temperature scenarios (Atkins, De Paula, & Keeler, 2025). The bomb calorimeter utilized in this experiment, such as the Parr bomb calorimeter, operates under constant volume conditions whereby the sample combusts inside a sealed vessel, or “bomb” (Parr Instrument Company, 2013). This constant volume setting ensures no expansion work occurs, so changes in internal energy correspond solely to heat transfer. Since the bomb is not perfectly adiabatic and combustion is exothermic, heat flows into the surrounding water, causing a measurable temperature rise. Though perfect adiabatic isolation is unattainable, the system—comprising bomb and water—is treated as thermally isolated for practical purposes.

Importantly, the water surrounding the bomb is at constant pressure and free to expand, so the First Law does not eliminate work terms. Here, entropy becomes a useful state variable proxy for heat exchange (as mathematically defined below), accounting for pressure-volume work during heat transfer. This allows enthalpy—a state variable—to represent the measured energy changes, which can be further applied in Hess’s Law calculations for comprehensive  thermodynamic analysis under constant pressure conditions (Atkins, De Paula, & Keeler, 2025).Before moving on it should be noted that it is important to calibrate the bomb with a known species such as Benzoic acid before attempting to do any other experimental calculations.

Finally, given state variables can meaningfully be recorded and related under different conditions of state, we can also substantiate our experimental analysis through comparison to theoretical calculations using sites such as MolCalc and NIST Webbook.

## 4.3. Equations Implemented Throughout This Work

List of equations implemented throughout the course of the work:

**First Law of Thermodynamics:**

$\Delta U = q + w$

**Enthalpy Definition:**

$H = U + PV$

**Change in Enthalpy at Constant Pressure:**

$\Delta H = \Delta U + P\Delta V$

**Relationship between $\Delta H$ and $\Delta U$ for reactions involving gases:**

$\Delta H = \Delta U + \Delta n_{gas} RT$

**Hess's Law (Sum of Steps):**

$\Delta H_{total} = \Delta H_1 + \Delta H_2 + ... + \Delta H_n$

**Hess's Law (using Enthalpies of Formation):**

$\Delta H_c^\circ = \sum (\nu_p \Delta H_f^\circ (products)) - \sum (\nu_r \Delta H_f^\circ (reactants))$

**Heat Released in Calorimetry:**

$Q_{released} = C_{eq} \times \Delta T$

**Heat from Wire Combustion:**

$Q_{wire} = H_{fe} \times W_{wire\_burned}$

**Heat from Sample Combustion:**

$Q_{sample} = Q_{released} - Q_{wire}$

**Enthalpy of Combustion per Gram:**

$H_{comb\_per\_g} = \frac{Q_{sample}}{mass_{sample}}$

**Enthalpy of Combustion per Mole:**

$H_{comb\_per\_mol} = -\frac{Q_{sample}}{mass_{sample}} \times MW \times \frac{1}{1000}$ (converted to kJ/mol)

**Propagation of Uncertainty (General Form for $f = x \pm y$ or $f = xy$ or $f = x/y$):**

For $f = x \pm y$: $\sigma_f = \sqrt{\sigma_x^2 + \sigma_y^2}$

For $f = xy$ or $f = x/y$: $\frac{\sigma_f}{f} = \sqrt{\left(\frac{\sigma_x}{x}\right)^2 + \left(\frac{\sigma_y}{y}\right)^2}$

**Z-Score for comparing experimental and theoretical values:**

$Z = \frac{|Value_{experimental} - Value_{theoretical}|}{\sqrt{\sigma_{experimental}^2 + \sigma_{theoretical}^2}}$

```python
# @title
# Theoretical enthalpy of combustion calculations

dHf_CO2_g_lit = -393.51     # kJ/mol, ∆Hf CO2 gas
sigma_dHf_CO2_g_lit = 0.13  # kJ/mol, uncertainty in ∆Hf CO2 gas

dHf_H2O_g_lit = -241.826    # kJ/mol, ∆Hf H2O gas
sigma_dHf_H2O_g_lit = 0.040 # kJ/mol, uncertainty in ∆Hf H20 gas

dHvap_H2O = 40.7            # kJ/mol, ∆Hvap H20
sigma_dHvap_H2O = 0         # Assuming no uncertainty

dHf_O2_g_lit = 0.0          # kJ/mol, ∆Hf O2 gas
sigma_dHf_O2_g_lit = 0.0    # Assuming no uncertainty


# --- Calculation for the Hf liquid for H2O and its error propogation ---
dHf_H2O_l_lit = dHf_H2O_g_lit - dHvap_H2O

# Propagate uncertainty for difference:
sigma_dHf_H2O_l_lit_squared = (sigma_dHf_H2O_g_lit**2 + sigma_dHvap_H2O**2)
sigma_dHf_H2O_l_lit = math.sqrt(sigma_dHf_H2O_l_lit_squared)

# Using Hess's law to determine the Hc of all species

# ---- Benzoic Acid ----

# Using previous Hf and error
dHf_solid_ba_lit = -367.67 # kJ/mol
sigma_dHf_solid_ba_lit = 4.00 # kJ/mol

# Balanced Combustion Equation for Benzoic Acid: C7H6O2(s) + 15/2 O2(g) → 7 CO2(g) + 3 H2O(l)
n_CO2_ba = 7
n_H2O_l_ba = 3
n_ba = 1
n_O2_ba = 15/2

dHcomb_ba_theo_calc = (n_CO2_ba * dHf_CO2_g_lit + n_H2O_l_ba * dHf_H2O_l_lit) - (n_ba * dHf_solid_ba_lit + n_O2_ba * dHf_O2_g_lit)


# --- Error propagation for Theoretical ΔHcomb of Benzoic Acid ---
sigma_dHcomb_ba_theo_calc_squared = (
    (n_CO2_ba * sigma_dHf_CO2_g_lit)**2 +
    (n_H2O_l_ba * sigma_dHf_H2O_l_lit)**2 +
    (n_ba * sigma_dHf_solid_ba_lit)**2 +
    (n_O2_ba * sigma_dHf_O2_g_lit)**2
)

sigma_dHcomb_ba_theo_calc = math.sqrt(sigma_dHcomb_ba_theo_calc_squared)


# Print results
print(f"Theoretical ΔHcomb for Benzoic Acid: {dHcomb_ba_theo_calc:.2f} +/- {sigma_dHcomb_ba_theo_calc:.2f} kJ/mol")

# ---- Naphthalene ----

# Using previous Hf and error
dHf_solid_naph_lit = 106.18 # kJ/mol
sigma_dHf_solid_naph_lit = 0.25 # kJ/mol

# Balanced Combustion Equation for Naphthalene: C10H8(s) + 12 O2(g) -> 10 CO2(g) + 4 H2O(l)

n_CO2_naph = 10
n_H2O_l_naph = 4
n_naph = 1
n_O2_naph = 12

dHcomb_naph_theo_calc = (n_CO2_naph * dHf_CO2_g_lit + n_H2O_l_naph * dHf_H2O_l_lit) - (n_naph * dHf_solid_naph_lit + n_O2_naph * dHf_O2_g_lit)

# --- Error propagation for Theoretical ΔHcomb of Naphthalene ---
sigma_dHcomb_naph_theo_calc_squared = (
    (n_CO2_naph * sigma_dHf_CO2_g_lit)**2 +
    (n_H2O_l_naph * sigma_dHf_H2O_l_lit)**2 +
    (n_naph * sigma_dHf_solid_naph_lit)**2 +
    (n_O2_naph * sigma_dHf_O2_g_lit)**2
)

sigma_dHcomb_naph_theo_calc = math.sqrt(sigma_dHcomb_naph_theo_calc_squared)


# Print results
print(f"Theoretical ΔHcomb for Naphthalene: {dHcomb_naph_theo_calc:.2f} +/- {sigma_dHcomb_naph_theo_calc:.2f} kJ/mol")

# ---- Palmitic Acid ----

# Using previous Hf and error
dHf_solid_palm_lit = -884.00 # kJ/mol
sigma_dHf_solid_palm_lit = 47.89 # kJ/mol

# Balanced Combustion Equation for Palmitic Acid: C16H32O2(s) + 23 O2(g) -> 16 CO2(g) + 16 H2O(l)

n_CO2_palm = 16
n_H2O_l_palm = 16
n_palm = 1
n_O2_palm = 23


dHcomb_palm_theo_calc = (n_CO2_palm * dHf_CO2_g_lit + n_H2O_l_palm * dHf_H2O_l_lit) - (n_palm * dHf_solid_palm_lit + n_O2_palm * dHf_O2_g_lit)

# --- Error propagation for Theoretical ΔHcomb of Palmitic Acid ---
sigma_dHcomb_palm_theo_calc_squared = (
    (n_CO2_palm * sigma_dHf_CO2_g_lit)**2 +
    (n_H2O_l_palm * sigma_dHf_H2O_l_lit)**2 +
    (n_palm * sigma_dHf_solid_palm_lit)**2 +
    (n_O2_palm * sigma_dHf_O2_g_lit)**2
)

sigma_dHcomb_palm_theo_calc = math.sqrt(sigma_dHcomb_palm_theo_calc_squared) # 1 sigma uncertainty


# Print results
print(f"Theoretical ΔHcomb for Palmitic Acid: {dHcomb_palm_theo_calc:.2f} +/- {sigma_dHcomb_palm_theo_calc:.2f} kJ/mol")

# ---- α-d-Glucose ----

# Using previous Hf and error
dHf_solid_glucose_lit = -1469.00 # KJ/mol
sigma_dHf_solid_glucose_lit = 25.36 # KJ/mol

# Balanced Combustion Equation for α-d-Glucose: C6H12O6(s) + 6 O2(g) -> 6 CO2(g) + 6 H2O(l)

n_CO2_glucose = 6
n_H2O_l_glucose = 6
n_glucose = 1
n_O2_glucose = 6

dHcomb_glucose_theo_calc = (n_CO2_glucose * dHf_CO2_g_lit + n_H2O_l_glucose * dHf_H2O_l_lit) - (n_glucose * dHf_solid_glucose_lit + n_O2_glucose * dHf_O2_g_lit)

# --- Error propagation for Theoretical ΔHcomb of α-d-Glucose ---
sigma_dHcomb_glucose_theo_calc_squared = (
    (n_CO2_glucose * sigma_dHf_CO2_g_lit)**2 +
    (n_H2O_l_glucose * sigma_dHf_H2O_l_lit)**2 +
    (n_glucose * sigma_dHf_solid_glucose_lit)**2 +
    (n_O2_glucose * sigma_dHf_O2_g_lit)**2
)

sigma_dHcomb_glucose_theo_calc = math.sqrt(sigma_dHcomb_glucose_theo_calc_squared)


# Print results
print(f"Theoretical ΔHcomb for α-d-Glucose: {dHcomb_glucose_theo_calc:.2f} +/- {sigma_dHcomb_glucose_theo_calc:.2f} kJ/mol")
```

```python
# @title
import numpy as np
import pandas as pd
import scipy.stats as stats

# ---- Number of points to use at start and end ----
Number_Points = 10

# ---- Select first and last points ----
initial_points_temp = temp_data_benzoic_acid.head(Number_Points)['Temperature (C)']
final_points_temp = temp_data_benzoic_acid.tail(Number_Points)['Temperature (C)']
initial_points_time = temp_data_benzoic_acid.head(Number_Points)['Time (s)']
final_points_time = temp_data_benzoic_acid.tail(Number_Points)['Time (s)']

# ---- Linear fits: y = m*x + b ----
mi, bi = np.polyfit(initial_points_time, initial_points_temp, 1)
mf, bf = np.polyfit(final_points_time, final_points_temp, 1)

# ---- Reference time ----
t_ref = 375  # seconds

# ---- Extrapolated temperatures ----
T_initial_ref = mi * t_ref + bi
T_final_ref = mf * t_ref + bf

# ---- ΔT ----
delta_T = T_final_ref - T_initial_ref


# ---- Estimate random error in ΔT from plateau standard deviations ----
sigma_initial = initial_points_temp.std()
sigma_final = final_points_temp.std()

# Estimate the standard error in ΔT by propagating the standard deviations of the plateaus.
sigma_delta_T = np.sqrt(sigma_initial**2 + sigma_final**2)


# ---- 95% confidence interval using t-value ----
df_total = len(initial_points_temp) + len(final_points_temp)

# For a two-sided 95% confidence interval for ∆T
t_value = stats.t.ppf(0.975, df_total)
delta_95 = t_value * sigma_delta_T


# ---- Print results ----
print(f"ΔT = {delta_T:.4f} °C")
print(f"Propagated σ_ΔT = {sigma_delta_T:.6f} °C")
print(f"ΔT ± 95% CI = {delta_T:.4f} ± {delta_95:.4f} °C")
```

```python
# @title
import pandas as pd
import numpy as np

# Assuming df_comparison is your DataFrame with experimental and theoretical columns

# Calculate percent errors avoiding division by zero by adding a small epsilon
epsilon = 1e-10

df_percent_error = pd.DataFrame({
    "Substance": df_comparison["Substance"],
    "Percent Error Enthalpy Combustion (%)": (np.abs(df_comparison["Experimental Enthalpy Combustion (kJ/mol)"] - df_comparison["Theoretical Enthalpy Combustion (kJ/mol)"]) / (np.abs(df_comparison["Theoretical Enthalpy Combustion (kJ/mol)"]) + epsilon)) * 100,
    "Percent Error Enthalpy Formation (%)": (np.abs(df_comparison["Experimental Enthalpy Formation (kJ/mol)"] - df_comparison["Theoretical Enthalpy Formation (kJ/mol)"]) / (np.abs(df_comparison["Theoretical Enthalpy Formation (kJ/mol)"]) + epsilon)) * 100,
})

# Optional: round values for readability
df_percent_error["Percent Error Enthalpy Combustion (%)"] = df_percent_error["Percent Error Enthalpy Combustion (%)"].round(2)
df_percent_error["Percent Error Enthalpy Formation (%)"] = df_percent_error["Percent Error Enthalpy Formation (%)"].round(2)

print("Percent Errors Between Experimental and Theoretical Data:")
print(df_percent_error.to_string(index=False))
```

```python
# @title
import numpy as np
import pandas as pd

def two_sided_z_test(df, value_exp, value_theo, uncert_exp, uncert_theo, alpha=0.05):
    results = []
    z_critical = abs(np.round(np.percentile(np.random.normal(0,1,1000000), [100*alpha/2, 100*(1-alpha/2)])[1],2))

    for _, row in df.iterrows():
        diff = row[value_exp] - row[value_theo]
        combined_uncert = np.sqrt(row[uncert_exp]**2 + row[uncert_theo]**2)
        z_score = abs(diff) / combined_uncert if combined_uncert > 0 else np.nan
        significant = z_score > z_critical
        results.append({
            "Substance": row["Substance"],
            "Z-Score": round(z_score, 2),
            "Critical Z": z_critical,
            "Significant Difference": significant
        })
    return pd.DataFrame(results)

# Example usage for enthalpy of combustion:
z_test_combustion = two_sided_z_test(
    df_comparison,
    "Experimental Enthalpy Combustion (kJ/mol)",
    "Theoretical Enthalpy Combustion (kJ/mol)",
    "Experimental Combustion Uncertainty (kJ/mol)",
    "Theoretical Combustion Uncertainty (kJ/mol)"
)

# Example usage for enthalpy of formation:
z_test_formation = two_sided_z_test(
    df_comparison,
    "Experimental Enthalpy Formation (kJ/mol)",
    "Theoretical Enthalpy Formation (kJ/mol)",
    "Experimental Formation Uncertainty (kJ/mol)",
    "Theoretical Formation Uncertainty (kJ/mol)"
)

print("Two-sided Z-test results for Enthalpy of Combustion:")
print(z_test_combustion.to_string(index=False))
print()
print("Two-sided Z-test results for Enthalpy of Formation:")
print(z_test_formation.to_string(index=False))

```

**Focusing on the difference between Palmitic Acid and α-d-Glucose:**

When seeking to describe and analyze experimentally determined enthalpy differences, its important to consider the underlying molecular structures. Major functional group differences and aromaticity,  amongst other factors well anticipated factors, endow molecules with varying energies. More fine alteractions can also be detected such as ring strain, as suggested by (Martin, Davidson, & Ball, 2016), where differnces between side chain placement and ring strain (cyclization) were deamed to yield calculable enthalpy differences. Furthermore, the concept of fine differences in structure yielding measurable differences in enthalpy is also well demonstrated in (Demaison & Aston, 1998), where two aromatic structural isomers, naphthalene and azulene, have small yet measureable differences in their heats of combustion. Far from isomers, α-d-Glucose, a sugar, and Palmitic Acid, a fatty acid, have largely different molecular structures. α-d-Glucose is a six carbon sugar that commonly adopts a pyranose cyclic structure. Moreover, being a sugar it is highly oxidized with five hydrozyl groups and an aldehyde or ether when linear or cyclic, respectively. On the other hand, Palmitic acid is a sixteen carbon fatty acid that, though carboxyl group containing, is far more reduced. These two well known and studied chemical species have very predictable enthalpies as they're members of the glycolytic and TCA cycles. (Li et al., 2010) provides a direct point of comparison between the two in which points throughout the progressive oxidation of the molecules can be traced and, thus, their decreased capacity to be used as an energy source for ATP production or for other energetic demands the cell has.
